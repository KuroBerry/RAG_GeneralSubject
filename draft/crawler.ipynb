```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.formula1.com/en/results/2025/drivers"

response = requests.get(url) # This line is just to ensure the URL is reachable and the request is made.
response
```

```python
if response.status_code == 200: # Check if the request was successful
    soup = BeautifulSoup(response.content, 'html.parser')
    table = soup.find('table', class_='f1-table f1-table-with-data w-full')  # Find the table with the class 'resultsarchive-table'
    tr = table.find_all('tr')  # Find all rows in the tbody

    for row_id, row in enumerate(tr):
        colums = row.find_all('td')  # Find all columns in the row
        # print(f'row:{row_id} {colums}')
        if colums:
            position = colums[0].text.strip() # Get the position from the first column
            driver_name = colums[1].text.strip()[:-3]  # Get the driver name from the second column
            nationality= colums[2].text.strip()
            car = colums[3].text.strip()
            points = colums[4].text.strip()
            
            print(f'Position: {position}, \nDriver: {driver_name}, \nNationality: {nationality}, \nCar: {car}, \nPoints: {points}')
            print('-' * 40)  # Print a separator line
```
