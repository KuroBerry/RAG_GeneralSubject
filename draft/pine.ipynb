```python
def embed_text(text):
    results = pc.inference.embed(
    model="llama-text-embed-v2",
    inputs=text,
    parameters={
        "input_type": "passage",
        "truncate": "END"
    }
    )
    return results[0]['values']
```

```python
embed_texts = embed_text(records[0]["text"])
embed_texts
```

```python
processed_records = []

for rec in records:
    processed_record = {
        "id": rec["id"],
        "values": embed_text(rec["text"]),
        "metadata": {
            "chunk_text": rec["text"],
            "category": rec["category"]
        },
    }
    processed_records.append(processed_record)
```
